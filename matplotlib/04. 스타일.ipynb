## matplotlib (https://matplotlib.org/stable/index)

- 다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] ="Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False
```

```python
x = [1,2,3]
y = [2,4,8]
```

#### 1. 선 굵기 : linewidth

```python
# plot(x축데이터,y축데이터,.....)
# label="데이터" : 범례 지정
# linewidth=10

g=plt.plot(x,y,linewidth=10)
```

#### 2. 마커(https://matplotlib.org/stable/api/markers_api.html)

```python
# plot(x축데이터,y축데이터,.....)
# label="데이터" : 범례 지정
# linewidth=10
# marker=""

g=plt.plot(x,y,linewidth=2,marker="o")
```

```python
# plot(x축데이터,y축데이터,.....)
# label="데이터" : 범례 지정
# linewidth=10
# marker="마커모양"

g=plt.plot(x,y,linewidth=2,marker="o",markersize=10)
```

```python
plt.plot([4,5,6],marker="H")
plt.plot([3,4,5],marker="d")
plt.plot([2,3,4],marker="x")
plt.plot([1,2,3],marker=11)
plt.plot([0,1,2],marker="$Z$")
plt.show()
```

```python
g=plt.plot(x,y,marker="o",markersize=20,markeredgecolor='red', markerfacecolor="yellow")
```

#### 3. 선 모양 : linestyle(https://matplotlib.org/stable/gallery/lines_bars_and_markers/linestyles.html)

```python
g = plt.plot(x, y, linestyle=":",linewidth=5)
```

```python
g = plt.plot(x, y, linestyle="--",linewidth=5)
```

```python
g = plt.plot(x, y, linestyle="-.",linewidth=5)
```

#### 4. 선 색상 : color(https://matplotlib.org/3.5.0/gallery/color/named_colors.html)

```python
g = plt.plot(x, y, linestyle="--",linewidth=5,color="red")
```

```python
g = plt.plot(x, y, linestyle="--",linewidth=5,color="r")
```

```python
g = plt.plot(x, y, linestyle="--",linewidth=5,color="b")
```

```python
g = plt.plot(x, y, linestyle="--",linewidth=5,color="g")
```

#### 5. 포맷

```python
g=plt.plot(x,y,'o')
```

```python
g=plt.plot(x,y,'ro')
```

```python
g=plt.plot(x,y,'ro--') # marker="o", color="r", linestyle="--"
```

```python
g=plt.plot(x,y,'bv:')
```

#### 6. 축약(https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html)

```python
g=plt.plot(x,y,marker="o",ms=20,mec='red', mfc="yellow",ls=":")
```

#### 7. 투명도

```python
g=plt.plot(x,y,marker="o",ms=20,mec='red', mfc="yellow",ls=":",alpha=0.3)
```

#### 8. 그래프 크기

```python
plt.figure(figsize=(5,5))
g=plt.plot(x,y)
```

#### 9. 배경색

```python
plt.figure(figsize=(10,5),facecolor="yellow")
g=plt.plot(x,y)
```
